There is three variants of QuickSort within one approach:  
We pick QuickSort-first method, where **use first element as pivot**.  
**If** pivot is **not first** element we just **swap** the pivot and first element. After it we have first-element case.  

it can be done QuickSort for last (see in '__Prog3_part2 (last).ipynb__') or for some another element (suddenly not implemented for the end in '**Prog3_part3 (median).ipynb**' and previous notebooks). But it give some technical problem.

# First 

```python
def Partition(A, l, r):
    pivot, p_ind = A[l], l
    i = l + 1
    for j in range(l + 1, r):
        if A[j] < pivot:
            A[j], A[i] = A[i], A[j]
            i += 1
    A[l], A[i - 1] = A[i - 1], A[l]
    return i - 1 


def QuickSort_first(A, l, r):
    global arr_count
    arr_count.append(len(A[l:r-1]))
    global count
    count += len(A[l:r-1])
    if l >= r:
        return None
    pivot = Partition(A, l, r)
    QuickSort_first(A, l, pivot)
    QuickSort_first(A, pivot + 1, r)
        
# arr = [1, 3, 4, 2]
# QuickSort_median(arr, 0, len(arr))
# arr

# arr = list(reversed(range(25)))
# arr = [8,171, 1, 123, 10]
# QuickSort_first(arr, 0, len(arr))
# arr
```

# Last 

```python
def Partition_last(A, start_ind, end_ind):
    pivot, p_ind = A[end_ind - 1], end_ind - 1

    A[p_ind], A[start_ind] = A[start_ind], pivot

    i = start_ind + 1
    for j in range(start_ind + 1, end_ind):
        if A[j] < pivot:
            A[j], A[i] = A[i], A[j]
            i += 1

    A[start_ind], A[i - 1] = A[i - 1], A[start_ind]
    return i - 1 

def QuickSort_last(A, start_ind, end_ind):
    global arr_count
    arr_count.append(len(A[start_ind:end_ind-1]))
    global count
    count += len(A[start_ind:end_ind-1])
    
    if start_ind < end_ind:
        pivot = Partition_last(A, start_ind, end_ind)
        QuickSort_last(A, start_ind, pivot)
        QuickSort_last(A, pivot + 1, end_ind)
        
# arr = [1, 3, 4, 2]
# QuickSort_median(arr, 0, len(arr))
# arr

# arr = list(reversed(range(25)))
# arr = [8,171, 1, 123, 10]
# QuickSort_last(arr, 0, len(arr))
# arr[:15]
```

# Median

```python
def median(arr, start_ind, end_ind):
    mid = (start_ind + end_ind-1)//2
    a = arr[start_ind]
    b = arr[mid]
    c = arr[end_ind - 1]
    if (a <= b <= c) | (c <= b <= a):
        return b, mid
    if (a <= c <= b) | (b <= c <= a):
        return c, end_ind - 1
    return a, start_ind
        

#A method to partition around the median
def Partition_median(A, start_ind, end_ind):
    pivot, p_ind = median(A, start_ind, end_ind)

    A[p_ind], A[start_ind] = A[start_ind], pivot

    i = start_ind + 1
    for j in range(start_ind + 1, end_ind):
        if A[j] < pivot:
            A[j], A[i] = A[i], A[j]
            i += 1

    A[start_ind], A[i - 1] = A[i - 1], A[start_ind]
    return i - 1 

def QuickSort_median(A, start_ind, end_ind):
    global arr_count
    arr_count.append(len(A[start_ind:end_ind-1]))
    global count
    count += len(A[start_ind:end_ind-1])
    
    if start_ind < end_ind:
        pivot = Partition_median(A, start_ind, end_ind)
        QuickSort_median(A, start_ind, pivot)
        QuickSort_median(A, pivot + 1, end_ind)
        
# arr = [1, 3, 4, 2]
# QuickSort_median(arr, 0, len(arr))
# arr

# arr = list(reversed(range(25)))
# arr = [8,171, 1, 123, 10]
# QuickSort_median(arr, 0, len(arr))
# arr
```
